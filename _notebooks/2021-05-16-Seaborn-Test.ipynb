# Package Installs

```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

# Data

```python
autos = pd.read_csv("./2021-05-16-BDC-Files/autos.csv")

autos.head()
```

```python
ax = sns.boxplot(
    data = autos.loc[autos["price"] <= 15000],
    y = "price"
)

ax.set(
    title = "Boxplot of Used Car Prices",
    xlabel = 'Used Cars',
    ylabel = 'Price',
)

ax.grid(True)

plt.show()
```
